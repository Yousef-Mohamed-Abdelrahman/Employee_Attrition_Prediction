```python
plt.figure(figsize=(10, 5))

ax = sns.countplot(data=data,x='Attrition',hue='Attrition',palette='Set2',legend=False)

total = len(data)

for p in ax.patches:
    count = int(p.get_height())
    percentage = 100 * count / total
    ax.annotate(f'{count}\n({percentage:.1f}%)',(p.get_x() + p.get_width() / 2, count),ha='center',va='bottom')

plt.title('Employee Attrition Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Attrition Status')
plt.ylabel('Employee Count')

sns.despine()
plt.tight_layout()
plt.show()
```

```python
attrition_rate = (data['Attrition'].value_counts(normalize=True)* 100)

print(attrition_rate)
```

```python
plt.figure(figsize=(8, 5))

ax = sns.boxplot(data=data,x='Attrition',y='Age',hue='Attrition',legend=False)


sns.stripplot(data=data,x='Attrition',y='Age',color='black',alpha=0.4,size=3)

plt.title('Age Distribution by Attrition Status', fontsize=14, fontweight='bold')
plt.xlabel('Attrition')
plt.ylabel('Age')

sns.despine()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))

sns.boxplot(
    data=data,
    x='Attrition',
    y='MonthlyIncome',
    hue='Attrition',
    palette='Set2',
    legend=False
)

sns.stripplot(
    data=data,
    x='Attrition',
    y='MonthlyIncome',
    color='black',
    alpha=0.3,
    size=2
)

plt.title('Monthly Income by Attrition Status')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')

sns.despine()
plt.tight_layout()
plt.show()
```

```python
pd.crosstab(data['OverTime'],data['Attrition'])
```

```python
prop_df = (
    pd.crosstab(data['OverTime'], data['Attrition'], normalize='index')
    .reset_index()
    .melt(id_vars='OverTime',
          var_name='Attrition',
          value_name='Proportion')
)

plt.figure(figsize=(8, 5))

ax = sns.barplot(
    data=prop_df,
    x='OverTime',
    y='Proportion',
    hue='Attrition',
    palette='Set2'
)

# Percentage labels
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f')

plt.title('Attrition Rate by Overtime Status')
plt.xlabel('Overtime')
plt.ylabel('Proportion')
plt.gca().yaxis.set_major_formatter(
    plt.matplotlib.ticker.PercentFormatter(1)
)

plt.tight_layout()
plt.show()
```

```python
prop_df = (
    pd.crosstab(data['JobSatisfaction'], data['Attrition'], normalize='index')
    .reset_index()
    .melt(
        id_vars='JobSatisfaction',
        var_name='Attrition',
        value_name='Proportion'
    )
)

plt.figure(figsize=(8, 5))

ax = sns.barplot(
    data=prop_df,
    x='JobSatisfaction',
    y='Proportion',
    hue='Attrition',
    palette='Set2'
)

for container in ax.containers:
    ax.bar_label(container, fmt='%.1f')

ax.yaxis.set_major_formatter(plt.matplotlib.ticker.PercentFormatter(1))

plt.title('Attrition Rate by Job Satisfaction')
plt.xlabel('Job Satisfaction')
plt.ylabel('Percentage of Employees')

plt.tight_layout()
plt.show()
```

```python
attrition_rate = (
    pd.crosstab(
        data['WorkLifeBalance'],
        data['Attrition'],
        normalize='index'
    )['Yes']
    .reset_index()
)

plt.figure(figsize=(8, 5))

ax = sns.barplot(
    data=attrition_rate,
    x='WorkLifeBalance',
    y='Yes',
    color='salmon'
)

ax.bar_label(
    ax.containers[0],
    fmt='%.1f'
)

ax.yaxis.set_major_formatter(plt.matplotlib.ticker.PercentFormatter(1))

plt.title('Attrition Rate by Work-Life Balance')
plt.xlabel('Work-Life Balance')
plt.ylabel('Attrition Rate')

plt.tight_layout()
plt.show()
```
