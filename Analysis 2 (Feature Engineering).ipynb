# Feature Engineering


```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Datasets/WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
data.head()
```

```python
drop_cols = ['EmployeeCount','StandardHours','EmployeeNumber','Over18']

data_cleaned = data.drop(columns=drop_cols)
```

```python
data_cleaned.head()
```

```python
data_cleaned['AgeGroup'] = pd.cut(data_cleaned['Age'],bins=[18,25,35,45,60],labels=['18-25','26-35','36-45','46-60'])
```

```python
pd.crosstab(data_cleaned['AgeGroup'],data_cleaned['Attrition'],normalize='index')*100
```

```python
data_cleaned['IncomeGroup'] = pd.qcut(data_cleaned['MonthlyIncome'],q=4)
```

```python
pd.crosstab(data_cleaned['IncomeGroup'],data_cleaned['Attrition'],normalize='index')*100
```

```python
pd.crosstab(data_cleaned['OverTime'],data_cleaned['JobRole'],values=(data_cleaned['Attrition'] == 'Yes'),aggfunc='mean')*100
```

```python
data_cleaned['IsYoungEmployee'] = np.where(data_cleaned['Age'] < 30,1,0)
```

```python
data_cleaned['LowIncome'] = np.where(data_cleaned['MonthlyIncome'] < data_cleaned['MonthlyIncome'].median(),1,0)
```

```python
income_median = data_cleaned['MonthlyIncome'].median()

print(income_median)
```

```python
data_cleaned['LongDistance'] = np.where(data_cleaned['DistanceFromHome'] > 10,1,0)
```

```python
data_cleaned['PromotionDelay'] = np.where(data_cleaned['YearsSinceLastPromotion'] >= 5,1,0)
```

```python
data_cleaned['HighRiskOvertime'] = np.where((data_cleaned['OverTime'] == 'Yes') &
(data_cleaned['JobRole'].isin(['Sales Representative','Laboratory Technician','Human Resources'])),1,0)
```

```python
data_cleaned.head()
```

```python
cols_to_drop = ['AgeGroup','IncomeGroup']

data_cleaned.drop(columns=cols_to_drop,inplace=True)
data_cleaned.head()
```

```python
data_cleaned.to_csv("Datasets/processed_hr_data.csv",index=False)
```

```python
data_encoded = pd.get_dummies(data_cleaned,drop_first=True,dtype=int)
```
